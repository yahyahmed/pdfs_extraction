```python
#unzipping the file
import zipfile
import os

# If you uploaded the ZIP file directly to Colab
zip_path = 'KCS110000.zip'  # Replace with the name of your uploaded ZIP file

# OR if the ZIP file is in Google Drive, adjust the path accordingly
# zip_path = '/content/drive/My Drive/path_to_your_zip_file.zip'

# Unzip the file to a specified directory
extract_to_dir = 'unzipped_files'  # The directory where you want to extract the files
os.makedirs(extract_to_dir, exist_ok=True)  # Create the directory if it doesn't exist

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_dir)

# Now, all files are extracted to 'unzipped_files' directory
```

```python
#processing the PDF
import fitz  # PyMuPDF

def convert_pdf_to_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Iterate over all extracted PDF files
for filename in os.listdir(extract_to_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(extract_to_dir, filename)
        text = convert_pdf_to_text(pdf_path)
        print(f"Text from {filename}:")
        print(text[:500])  # Print the first 500 characters to check
        # You can save this text to a file or process it further as needed
```

```python
import fitz  # PyMuPDF

def convert_pdf_to_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Directory where the extracted text files will be saved
output_dir = r'C:\Users\exone\Desktop\doAZ.ai\extracted_text'
os.makedirs(output_dir, exist_ok=True)

# Iterate over all extracted PDF files
for filename in os.listdir(extract_to_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(extract_to_dir, filename)
        text = convert_pdf_to_text(pdf_path)

        # Define the output text file name
        output_file_name = os.path.splitext(filename)[0] + '.txt'
        output_file_path = os.path.join(output_dir, output_file_name)

        # Write the extracted text to a new text file
        with open(output_file_path, 'w', encoding='utf-8') as text_file:
            text_file.write(text)

        print(f"Extracted text from {filename} has been saved to {output_file_path}")
```

```python
import shutil

destination_dir = r'C:\Users\exone\Desktop\doAZ.ai\extracted_text' #target directory
os.makedirs(destination_dir, exist_ok=True)  # Ensure the target directory exists

for filename in os.listdir(output_dir):
    if filename.endswith('.txt'):
        shutil.move(os.path.join(output_dir, filename), os.path.join(destination_dir, filename))
```
